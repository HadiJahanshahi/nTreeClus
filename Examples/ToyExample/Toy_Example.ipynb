# The small list of sequences

```python
my_list = ['evidence','evident','provide','unconventional','convene']
```

# Fitting nTreeClus on the list

```python
import nTreeClus
model      = nTreeClus.nTreeClus(my_list, n=None, ntree=10, method="All", verbose=0)
model.nTreeClus()
```

# Taking a look at the output of the model

```python
model.output()
```

# Plotting the Dendrogram for Decision Tree version of nTreeClus 

```python
from matplotlib import pyplot as plt
fig, ax = model.plot('DT', my_list, linkage_method= 'ward', rotation=0)
plt.show()
```

# Plotting the Dendrogram for Random Forest version of nTreeClus 

```python
from matplotlib import pyplot as plt
fig, ax = model.plot('RF', my_list, linkage_method= 'ward', rotation=0)
plt.show()
```

As the figures demonstrate, the output of RF and DT implementation of nTreeClus is very similar, and the method is robust to method selection.

# Plotting the Dendrogram for Position-sensitive version of nTreeClus

```python
from matplotlib import pyplot as plt
fig, ax = model.plot('RF_position', my_list, linkage_method= 'ward', rotation=0)
plt.show()
```

When we use position-based nTreeClus, the difference between `provide` and the other two words with the root of `vid` will become more significant. The reason is that `vide` in `provide` comes toward the end, whereas it is closer to the beginning in `evident` and `evidence`.
